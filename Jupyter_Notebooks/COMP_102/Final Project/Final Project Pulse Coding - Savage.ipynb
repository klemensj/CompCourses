```python
import numpy as np
import pandas as pd

NOA_control = pd.read_csv('NOA_control.txt', sep ='\t', skiprows=6, names = ["interval","data"])
NOA_scared = pd.read_csv('NOA_scared.txt', sep ='\t', skiprows=6, names = ["interval","data"])
TYLER_control = pd.read_csv('TYLER_control.txt', sep ='\t', skiprows=6, names = ["interval","data"])
TYLER_scared = pd.read_csv('TYLER_scared.txt', sep ='\t', skiprows=6, names = ["interval","data"])
```

```python
NOA_control.head()
NOA_scared.head()
TYLER_control.head()
TYLER_scared.head()
```

```python
FearData = pd.concat([NOA_control['data'], NOA_scared['data'], TYLER_control['data'], TYLER_scared['data']], ignore_index = True, axis=1,)
```

```python
FearData.columns=["NOA_control","NOA_scared","TYLER_control","TYLER_scared"]
FearData.head()
```
