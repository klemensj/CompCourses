```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 


NOA_Control = pd.read_csv('desktop/NOA_control.txt', sep = '\t', skiprows= 7, names = ["interval","data"])
NOA_Scared = pd.read_csv('desktop/NOA_Scared.txt', sep = '\t', skiprows= 7, names = ["interval","data"])
TYLER_control = pd.read_csv('desktop/TYLER_control.txt', sep = '\t', skiprows= 7, names = ["interval","data"])
TYLER_Scared = pd.read_csv('desktop/TYLER_Scared.txt', sep = '\t', skiprows= 7, names = ["interval","data"])
NOA_Control.head()
NOA_Scared.head()
TYLER_control.head()
TYLER_Scared.head()
```

```python
NOA_Control ['interval']= pd.to_numeric (NOA_Control ['interval'], errors = 'coerce')
NOA_Control['data'] = pd.to_numeric(NOA_Control['data'], errors = 'coerce')

NOA_Control.plot.line(x="interval" , y="data" , title = "NOA Control", color="gray")
NOA_Scared.plot.line(x="interval" , y="data" , title = "Noa Scared", color ="grey")
#FearData= pd.concat([NOA_Control['data'],NOA_Scared['data'],TYLER_control['data'],TYLER_Scared['data']], ignore_index= True, axis=1)
#FearData.columns= ["NOA_Control","NOA_Scared","TYLER_control","TYLER_Scared" ]

```
