```python
import pandas as pd #don't change this line

import numpy as np

import matplotlib.pyplot as mpl



Noa_Control = pd.read_csv('NOA_control.txt', sep = '\t', names= ["interval" ,"control data"])
Noa_Scared = pd.read_csv('NOA_scared.txt', sep = '\t', names= ["interval", "scared data"])
Tyler_Control = pd.read_csv('TYLER_control.txt', sep = '\t', names= ["interval" , "control data"])
Tyler_Scared = pd.read_csv('TYLER_scared.txt', sep = '\t', names= ["interval" , "scared data"])


 
```

```python
Noa_Control.head(15)
```

```python
Noa_Scared.head(15)
```

```python
Noa_Control['interval'] = pd.to_numeric(Noa_Control['interval'],errors='coerce')   
Noa_Control['control data'] = pd.to_numeric(Noa_Control['control data'], errors='coerce')

#For some reason, NOA.control data is not stored as numerical values even though they are numbers, above code converts them. 
#The other files seem to be stored as numerical values and I didn't have this issue. 

Noa_Control.plot.line(x="interval" , y="control data" , color = "gray", title= "Noa Control")

Noa_Scared.plot.line(x="interval" , y="scared data", color = "chartreuse", title = "Noa Scared")
```

```python
# here I combined the data into one data frame to prepare for a combined graph using the common "interval" column

all_data_noa = pd.merge(Noa_Control,
                 Noa_Scared[['interval','scared data']],
                 on= 'interval')
all_data_noa.head(15)
```

```python
noa_graph = all_data_noa.plot(x="interval", y=["control data","scared data"], kind="line", title = "Noa's Combined Data", color=['gray', 'chartreuse'] )
from pylab import rcParams
rcParams['figure.figsize'] = 10,5


#I slightly altered the code for this, and changed color, line width, made the graph wider, but don't totally understand how 
```
