```python
import pandas as pd
import numpy as np
from scipy import stats 


N = 1000
calc_ave = 82
chem_ave = 85
biol_ave = 88

sd = 4

sex = []

univ = []

headers = ['univ','sex','calculus','chemistry','biology']

calc = np.random.normal(calc_ave, sd, N)
chem = np.random.normal(chem_ave, sd, N)
biol = np.random.normal(biol_ave, sd, N)

for i in range(1000):
    if biol[i] > 100:
        biol[i] = 100

for i in range(1000):
    if i % 2 == 0:
        sex.append('M')
    else:
        sex.append('F')
        
for i in range(1000):
    if i % 4 == 0:
        univ.append('ASU')
    elif i % 4 == 1:
        univ.append('BSU')    
    elif i % 4 == 2:
        univ.append('CSU')        
    elif i % 4 == 3:
        univ.append('DSU')  
        
print(len(sex))
print(len(univ))
print(len(calc))

df = pd.DataFrame(data = {'univ':univ, 'sex':sex, 'calc':calc, 'chem':chem, 'biol':biol})

df.describe()



```

```python
df.to_csv('data/test_aves.csv', index = False)
```
