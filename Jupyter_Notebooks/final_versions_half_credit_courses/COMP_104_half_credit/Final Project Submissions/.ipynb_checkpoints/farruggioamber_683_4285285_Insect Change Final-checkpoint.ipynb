```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats 

df = pd.read_csv('InsectAbundanceBiomassData.csv')
df.head()
data = number_no_na = df[df["Number"].notna()]
number_no_na.head()
```

```python
#Correlation Analysis 

x = data['Year']
y = data['Number']

thistuple = (stats.pearsonr(x,y))
print("r =", thistuple[0])
print("p-value =", thistuple[1])

#scatterplot

fig = plt.figure( figsize = (10,10))
ax = plt.axes()

ax.scatter = plt.scatter(x,y)

plt.ylabel('Number of Insects (1E6)', fontsize = 16)
plt.xlabel('Year', fontsize = 16)
plt.title('The Abundance of Insects Each Year Between 1959 and 2018', fontsize = 18, fontweight = 'bold')
```

```python
#Regression Analysis
x = data['Year']
y = data['Number']

mytuple = (stats.linregress(x,y))

print("R^2 = ", mytuple[2]**2)    
print("r = ", mytuple[2])
print("slope = ", mytuple[0])
print("intercept = ", mytuple[1])

#scatterplot
fig = plt.figure(figsize = (10,10))

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)  
line = slope*x + intercept 

plt.text(1940,4E6,('R squared = %f'% r_value**2), fontsize = 12)  
plt.text(1940,3.75E6,('p value = %f'% p_value), fontsize = 12)   

ax.scatter = plt.scatter(x,y)
plt.plot(x, line)

plt.ylabel('Number of Insects (1E6)', fontsize = 16)
plt.xlabel('Year', fontsize = 16)
plt.title('The Abundance of Insects Each Year Between 1959 and 2018', fontsize = 18, fontweight = 'bold')
```

```python
#Means Comparison

Trees = data[data["Stratum"].isin(['Trees'])]
Herblayer = data[data["Stratum"].isin(['Herb layer'])]

Treesmean = np.mean(Trees['Number'])
Herblayermean = np.mean(Herblayer['Number'])

Treessd = np.std(Trees['Number'])
Herblayersd = np.std(Herblayer['Number'])

print('Trees mean = ' + str(round(Treesmean,3)))
print('Trees standard deviation = ' + str(round(Treessd,3)))
print('Herb layer mean = ' + str(round(Herblayermean,3)))
print('Herb layer standard deviation = ' + str(round(Herblayersd, 3)))
print()

twosided = stats.ttest_ind(Trees['Number'],Herblayer['Number'])

print('t-statistic = '+ str(round(twosided.statistic,3)))
print('p value = '+ str(round(twosided.pvalue,5)))

plt.bar([1,2],[Treesmean, Herblayermean], yerr = [Treessd, Herblayersd], capsize = 10, color = 'blue', ecolor = 'red')

plt.ylim(0,350)

plt.annotate('t-statististic = -12.8', xy = (2,1), xytext = (1.15,320))
plt.annotate('Trees                                      Herb Layer', xy = (1,.05), xytext = (.85,285), fontweight = 'bold')
plt.title('The Average Number of Insects in Different Stratums', fontsize = 18, fontweight = 'bold')
```

```python
#Chi-Squared Analysis
df = pd.read_csv('InsectAbundanceBiomassData.csv')
data = np.array([[24.798, 90.293]])

data_df = pd.DataFrame(data, index = ["Abundance"], columns = ['Trees', 'Herb layer'])

chi2, p, dof, expected = stats.chi2_contingency(data_df)

print('Data:')
print()
print(data_df)
print()
print('Chi-square Results:')
print()
print('   chi-square = '+ str(round(chi2,4)))
print('   p value = '+ str(round(p,4)))
print('   degrees of freedom = '+ str(dof))
print('   expected values =')
print()
print()
print(pd.DataFrame(expected, index = ["Abundance"], 
                   columns = ['Trees', 'Herb layer']))
```
